```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Cluster 1 (class 0)
c1 = np.random.randn(150, 2) * [1.5, 1] + [0, 0]

# Cluster 2 (class 1)
c2 = np.random.randn(150, 2) * [1, 1.5] + [3, 3]

# Add some overlap noise
noise_0 = np.random.randn(50, 2) * 2 + [2, 1]
noise_1 = np.random.randn(50, 2) * 2 + [1, 2]

# Combine
X = np.vstack((c1, c2, noise_0, noise_1))
y = np.array([0]*150 + [1]*150 + [0]*50 + [1]*50)

# Shuffle
perm = np.random.permutation(len(X))
X = X[perm]
y = y[perm]

# Plot
plt.scatter(X[:,0], X[:,1], c=y, s=40)
plt.show()
```
