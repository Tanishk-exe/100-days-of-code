```python
import pandas as pd
```

first we are importing the data from which the model will be trained.

```python
heart=pd.read_csv('/content/heart-disease.csv')
heart.head()
```

Now, we have heart disease dataSet, Now we will divide the features into a and y where x is all the input variable or fields from which the model will predict.
Y will be target variable from which model will learn to predict the result

A feature is an input variable used to make a prediction.

A target is the output variable that the model predicts.

```python
x=heart.drop('target', axis=1)
y=heart['target']
y
```

From this we are importing Random forest classifier model and seeing the hyperparamters.

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier()
clf.get_params()
```

Here we have imported train test split from sklearn.model_selection which splits the data into training data and testing data
test size=0.2 means spliting data into 80% for training and 20% for testing

```python
from numpy import test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2)
```

```python
x_train.head()
```

Fitting the training data into the model

```python
clf.fit(x_train, y_train)
```

Now we will predict target or y variable using test data

```python
y_preds=clf.predict(x_test)
y_preds
```

Now we will evaluate the model using test data set of both x and y

```python
clf.score(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(y_test, y_preds))
```

```python
print(confusion_matrix(y_test, y_preds))
```

```python
print(accuracy_score(y_test, y_preds))
```

```python
import numpy as np
np.random.seed(42)
for i in range(10, 100, 10):
  clf2=RandomForestClassifier(n_estimators=i).fit(x_train, y_train)
  print(f'n_estimators is {i} and accuracy is {clf2.score(x_test, y_test) * 100}')
  print("")
```
