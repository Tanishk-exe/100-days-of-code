```python
heart.plot.hist(figsize=(10,30), subplots=True);
```

```python
over50=heart[heart['age'] >50 ];
over50.head()
```

```python
# pyplot method for quick visualization
over50.plot(kind="scatter", x='age', y='chol', c='target');
```

```python
#OO method for advanced Visualization
fig, ax=plt.subplots(figsize=(10,7))
over50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax);

#ax.set_xlim(40,100);
```

```python
from matplotlib.lines import lineStyles
fig, ax=plt.subplots(figsize=(10,6));
scatter=ax.scatter(x=over50['age'], y=over50['chol'], c=over50['target']);
ax.set(title="Heart Disease", xlabel="Age", ylabel="Cholestrol");
#Legend
ax.legend(*scatter.legend_elements(), title="Target");
#horizontal line
ax.axhline(over50['chol'].mean(), linestyle="--");
```

```python
fig, [ax1,ax2]=plt.subplots(nrows=2, ncols=1, figsize=(10,10), sharex=True)
b1=ax1.scatter(x=over50['age'], y=over50['chol'], c=over50['target']);
ax1.set(title="Cholestrol Level", ylabel="Cholestrol");
ax1.legend(*b1.legend_elements(), title="Target");
ax1.axhline(over50['chol'].mean(), linestyle="--");

b2=ax2.scatter(x=over50['age'], y=over50['thalach'], c=over50['target']);
ax2.set(title="Heart Rate Levels", xlabel="Age", ylabel="Haert Rate");
ax2.legend(*b2.legend_elements(), title="Target");
ax2.axhline(over50['thalach'].mean(), linestyle="--");

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");

fig.savefig("HeartDisease.png")
```

Heatmaps

```python
h1=sns.load_dataset('flights')
h1
```

```python
h1=h1.pivot(index="month", columns="year", values="passengers")
h1.head()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(h1);
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(h1, annot=True, fmt='d');
# fmt is format for annotaions d is decimal integer
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(h1, annot=True, fmt='d', linecolor='k', linewidths=2);
fig.savefig("Heatmap.png")
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(h1, annot=True, fmt='d', cmap='Reds', cbar=False);
#cmap gives color on the heatmap and cbar for showing bar or not
```
