```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

# **Quadratic Equation**

```python
np.random.seed(42)

x = np.linspace(-5, 5, 50).reshape(-1, 1)
y = x.flatten()**2 + np.random.normal(0, 3, size=50)
```

```python
plt.scatter(x,y);
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=4)
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly=PolynomialFeatures(degree=2)
xp=poly.fit_transform(x_train)
xtp=poly.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()

poly.fit(xp, y_train)
```

```python
lr.fit(xp, y_train)
```

```python
y_preds=lr.predict(xtp)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error

mean_absolute_error(y_test, y_preds)
```

```python
mean_squared_error(y_test, y_preds)
```

```python
root_mean_squared_error(y_test, y_preds)
```

# **Cubic Equation**

```python
x1 = np.linspace(-3, 3, 60).reshape(-1, 1)
y1 = x1.flatten()**3 + np.random.normal(0, 4, size=60)
```

```python
plt.scatter(x1,y1);
```

```python
x1_train, x1_test, y1_train, y1_test=train_test_split(x1,y1, test_size=0.33, random_state=4)
```

```python
lr1=LinearRegression()
poly1=PolynomialFeatures(degree=3)

xp1=poly1.fit_transform(x1_train)
xpt1=poly1.transform(x1_test)

poly1.fit(xp1, y1_train)
```

```python
lr1.fit(xp1, y1_train)
```

```python
pred=lr1.predict(xpt1)
```

```python
mean_absolute_error(y1_test, pred)
```

```python
mean_squared_error(y1_test, pred)
```

```python
root_mean_squared_error(y1_test, pred)
```
