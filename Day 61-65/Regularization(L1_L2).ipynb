```python
from sklearn.datasets import make_moons
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
```

```python
x,y=make_moons(n_samples=500, noise=0.05, random_state=5)
```

```python
plt.scatter(x[:,0], x[:,1], c=y)
```

```python
model=Sequential()
model.add(Dense(128, input_dim=2, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
adam=Adam(learning_rate=0.01)
model.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])
```

```python
log=model.fit(x,y, epochs=2000, validation_split=0.2, verbose=0)
```

```python
plt.plot(log.history['loss'])
plt.plot(log.history['val_loss'])
```

```python
plt.plot(log.history['accuracy'])
plt.plot(log.history['val_accuracy'])
```

```python
model1=Sequential()

model1.add(Dense(128,input_dim=2, activation='relu', kernel_regularizer=tensorflow.keras.regularizers.l2(0.03)))
model1.add(Dense(128, activation='relu', kernel_regularizer=tensorflow.keras.regularizers.l2(0.03)))
model1.add(Dense(1, activation='sigmoid'))
```

```python
adam=Adam(learning_rate=0.01)
model1.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])
```

```python
log2=model1.fit(x,y, validation_split=0.2, epochs=2000, verbose=0)
```

```python
plt.plot(log2.history['loss'])
plt.plot(log2.history['val_loss'])
```

```python
plt.plot(log2.history['accuracy'])
plt.plot(log2.history['val_accuracy'])
```
