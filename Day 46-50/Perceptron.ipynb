```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(42)

cgpa = np.random.uniform(5.0, 9.5, 100)
resume_score = np.random.uniform(3.0, 10.0, 100)

# Simple linear rule
placed = (cgpa + resume_score > 13).astype(int)

df = pd.DataFrame({
    "cgpa": cgpa,
    "resume_score": resume_score,
    "placed": placed
})

df
```

```python
sns.scatterplot(x=df['cgpa'], y=df['resume_score'], hue=df['placed'])
```

```python
from sklearn.linear_model import Perceptron
```

```python
p=Perceptron()
```

```python
x=df.drop('placed', axis=1)
y=df['placed']
```

```python
p.fit(x,y)
```
